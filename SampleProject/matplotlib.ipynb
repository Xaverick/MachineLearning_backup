```python
#1st method

fig = plt.figure() # creats a figure object
ax = fig.add_subplot() # add a subplot to the figure
ax.plot(x,y) # plot some data on the axes
plt.show()
```

```python
# 2nd method

fig = plt.figure() # create a figure object
ax = fig.add_axes([1,1,1,1]) # create an axes object in the figure
ax.plot(x,y) # plot the data
plt.show()
```

```python
# 3rd method (recommended)

fig, ax = plt.subplots()
ax.plot(x, y);
```

## Matplotlib Example workflow

```python
# 0. import matplotlib.pyplot and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#5. save & show (you save the whole figure)
fig.savefig("./images/sample-plot.png")
```

<h3> Two options for subplots </h3>

```python
# subplot option 1
fig, ((ax1,ax2) , (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize=(10,5))

#plot to each different axis
ax1.plot(x, x/2);   
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subolot option 2

fig ,ax = plt.subplots( nrows=2,
                        ncols=2,
                        figsize=(10,5))


#plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib oo method)

* When ploting something quickly, okay to use the pyplot method
* When plotting something advance use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50];
len(over_50)
```

```python
#Pyplot Method
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target');
```

```python
# OO Method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', 
            x='age',
            y='chol',
            c='target',
            ax=ax);

# ax.set_xlim([45,100]);
```

```python
#OO method completey from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                        y=over_50["chol"],
                        c=over_50["target"]
                        );

#customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
            linestyle="--");
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach

fig, (ax0 ,ax1) = plt.subplots(nrows=2,
                               ncols= 1,
                                figsize=(10, 10))


# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                    c=over_50["target"])


ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")


# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")


# Add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");


scatter = ax1.scatter(x=over_50["age"], 
                      y=over_50["thalach"],
                      c=over_50["target"])

ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age", 
        ylabel="Max Heart Rate");

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use("seaborn-whitegrid")


#OO method completey from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                        y=over_50["chol"],
                        c=over_50["target"],
                        cmap="winter"); #this changes the color scheme

#customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
            linestyle="--");
```

```python
# Customizing the y and the x axis limitations

# subplot of chol, age, thalach

fig, (ax0 ,ax1) = plt.subplots(nrows=2,
                               ncols= 1,
                                figsize=(10, 10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")

# Customize ax0

ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

# Change the y axis limits
ax0.set_xlim([50, 80])
ax0.set_ylim([100, 600])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");





# Add data to ax1

scatter = ax1.scatter(x=over_50["age"], 
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")

ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age", 
        ylabel="Max Heart Rate");

# change x1 axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
